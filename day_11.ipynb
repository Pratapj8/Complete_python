### 1. Retrieve the categories for a page
You can get the categories an article belongs to by setting the prop parameter to categories. This is useful for grouping or classifying pages.

```python
import requests
import json

S = requests.Session()

URL = "https://en.wikipedia.org/w/api.php"

PARAMS = {
    "action": "query",
    "format": "json",
    "titles": "Python_(programming_language)",
    "prop": "categories",
    "cllimit": "max"  # Get up to the maximum number of categories
}

HEADERS = {
    "User-Agent": "MyCustomApp/1.0 (myemail@example.com)"
}

response = S.get(url=URL, params=PARAMS, headers=HEADERS)
data = response.json()

# Extract and print categories
pages = data["query"]["pages"]
for page_id, page_data in pages.items():
    if "categories" in page_data:
        print(f"Categories for '{page_data['title']}':")
        for cat in page_data["categories"]:
            print(f"- {cat['title']}")
```

```python
import requests
import json

S = requests.Session()
URL = "https://en.wikipedia.org/w/api.php"

# Search for the term "Python programming"
PARAMS = {
    "action": "query",
    "format": "json",
    "list": "search",
    "srsearch": "Python programming"
}

response = S.get(url=URL, params=PARAMS)
data = response.json(0)

# Access and print the search results
search_results = data["query"]["search"]
for result in search_results:
    print(f"Title: {result['title']}")
    print(f"Snippet: {result['snippet']}")
    print("-" * 20)
```

```python
import requests

S = requests.Session()

URL = "https://en.wikipedia.org/w/api.php"

# The parameters for the API request
PARAMS = {
    "action": "query",
    "format": "json",
    "titles": "Python_(programming_language)",
    "prop": "extracts",
    "exlimit": "1",
    "explaintext": "1"
}

HEADERS = {
    "User-Agent": "MyCustomApp/1.0 (myemail@example.com)"
}

response = S.get(url=URL, params=PARAMS, headers=HEADERS)

# --- Debugging steps ---
print("Status Code:", response.status_code)
print("Response Text:")
print(response.text)

# Now, try to process as JSON
# You can put this in a try-except block
# to handle the error gracefully.
try:
    data = response.json()
    # Process the 'data' dictionary as needed
    print("\nSuccessfully parsed JSON data.")
    print(data)
except requests.exceptions.JSONDecodeError as e:
    print(f"\nJSONDecodeError caught: {e}")
    print("The response was not valid JSON.")
```

```python
import requests
import json
import pandas as pd

S = requests.Session()
URL = "https://en.wikipedia.org/w/api.php"

PARAMS = {
    "action": "query",
    "format": "json",
    "titles": "Python_(programming_language)",
    "prop": "extracts",
    "exlimit": "1",
    "explaintext": "1"
}

HEADERS = {
    "User-Agent": "MyCustomApp/1.0 (myemail@example.com)"
}

response = S.get(url=URL, params=PARAMS, headers=HEADERS)
data = response.json()

# Navigate the JSON structure to get the page data
pages = data["query"]["pages"]

# Get the first (and only) page from the response
page_id = next(iter(pages))
page_data = pages[page_id]

# Create a DataFrame from the extracted data
df = pd.DataFrame([{
    "title": page_data["title"],
    "extract": page_data["extract"]
}])

# Print the DataFrame
print(df)
```

```python
import pandas as pd

url = "https://en.wikipedia.org/wiki/Minnesota"
match_string = "Average daily maximum and minimum temperatures for selected cities in Minnesota"

try:
    # Attempt to read the HTML tables
    tables = pd.read_html(url, match=match_string)

    if tables:
        # Access the first table in the list and store it in a DataFrame
        df = tables[0]
        print("Table successfully loaded into a DataFrame.")
        print(df.head())
    else:
        print(f"No tables found matching '{match_string}'.")

except ValueError as e:
    print(f"An error occurred: {e}")
except IndexError:
    print(f"The table matching '{match_string}' could not be accessed.")
```
