```python
# Data type in csv
data = ("a,b,c,d\n"
        "1,2,3,4\n"
        "5,6,7,8\n"
        "9,10,11") # Due to 1 missing value it shows nan/missing value
```

```python
pd.read_csv(StringIO(data))
```

```python
df = pd.read_csv(StringIO(data))
```

```python
# d column is in float
df.info()
```

```python
# Change data type to object
df = pd.read_csv(StringIO(data),dtype="object")
df.info()
```

```python
df.isnull().sum()
```

```python
data = ("a,b,c,d\n"
        "1,2,3,4\n"
        "5,6,7,8\n"
        "9,10,11")
```

```python
# Change the data type to desired data types
df = pd.read_csv(StringIO(data),dtype={"a":int,"b":float,"c":int})
df.info()
```

```python
df.dtypes
```

```python
data = ("index,a,b,c\n"
        "4,apple,bat,5.7\n"
        "8,orange,cow,10")
```

```python
# Change the index column
pd.read_csv(StringIO(data),index_col=0)
```

```python
pd.read_csv(StringIO(data),index_col=1) # a column become index
```

```python
# use index cols and usecols
data = ("a,b,c\n"
        "4,apple,bat\n"
        "8,orange,cow,")
```

```python
pd.read_csv(StringIO(data),usecols=["a","b","c"],index_col=0) 
```
